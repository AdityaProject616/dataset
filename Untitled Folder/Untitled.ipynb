# Note :
1. Here, the following columns contain null values which need to be replaced with appropriate values
    1. trestbps (numerical)
    2. cholestrol (numerical)
    3. fbs
    4. restecg
    5. thalach (numerical)
    6. exang
    7. oldpeak (numerical)
    8. slope
    9. ca
    10. thal
2. Ideal strategy for replacing nunmerical values is to replace them with the mean value of the remaining, none null data.
3. For the categorical variables, modal value of the none null data can be considered for replacement.

# Now replacing categorical var with mode

```python
integrated_dataset.ca = integrated_dataset.ca.fillna(integrated_dataset['ca'].mode()[0])
integrated_dataset['ca'].unique()
```

```python
integrated_dataset['ca'] = integrated_dataset['ca'].astype(int)
```

```python
integrated_dataset['ca'].unique()
```

```python
integrated_dataset['slope'].unique()
```

```python
integrated_dataset['slope'] = integrated_dataset['slope'].fillna(integrated_dataset['slope'].mode()[0])
```

```python
integrated_dataset['slope'].unique()
```

```python
integrated_dataset['slope'] = integrated_dataset['slope'].astype(int)
```

```python
categorical_var = ['thal','exang','fbs','restecg']
for i in categorical_var:
    tempseries = integrated_dataset[str(i)]
    tempseries.fillna(tempseries.mode()[0],inplace=True)
    tempseries = tempseries.astype(int)
    integrated_dataset[str(i)] = tempseries
#     print(tempseries)
```

# Now replacing numerical var with mean

```python
numerical_vars = ['trestbps','cholestrol','thalach','oldpeak']
means = []
for i in numerical_vars:
    tempseries = integrated_dataset[i]
    tempseries.fillna(tempseries.mean(),inplace=True)
    means.append(tempseries.unique())
means
```

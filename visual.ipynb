```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Heart.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Ca'].fillna(df.Ca.mean(),inplace=True)
df['Thal'].fillna(df.Ca.mode()[0],inplace=True)
df.isnull().sum()
```

```python
sns.histplot(data=df,x='Age',kde=True, hue='Sex')
```

```python
sns.scatterplot(data=df,x='Age',y="Chol",hue='ChestPain')
```

```python
sns.barplot(data=df,x='ChestPain',y='Chol',hue='Sex')
```

```python
sns.lineplot(data=df,x='Age',y="Chol",hue='Sex')
```
